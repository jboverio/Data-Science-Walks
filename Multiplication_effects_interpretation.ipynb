```python
import numpy as np
import matplotlib.pyplot as plt

def demand_function(price, a, b):
    return a - b * price

def supply_function(price, c, d):
    return c + d * price

def find_equilibrium(a, b, c, d):
    price = (a - c) / (b + d)
    quantity = demand_function(price, a, b)
    return price, quantity

def rectangle_area(width, height):
    return width * height

a = 50  # Demand constant
b = 1   # Demand slope
c = 10  # Supply constant
d = 1   # Supply slope

eq_price, eq_quantity = find_equilibrium(a, b, c, d)
print("Equilibrium price:", eq_price)
print("Equilibrium quantity:", eq_quantity)

area = rectangle_area(eq_quantity, eq_price)
print("Area of the rectangle:", area)

prices = np.linspace(0, 50, 100)
demand = demand_function(prices, a, b)
supply = supply_function(prices, c, d)

plt.plot(prices, demand, label="Demand")
plt.plot(prices, supply, label="Supply")
plt.axvline(eq_price, color='red', linestyle='--', label='Equilibrium Price')
plt.axhline(eq_quantity, color='green', linestyle='--', label='Equilibrium Quantity')
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.legend()
plt.title("Supply and Demand Curves")
plt.show()
```
